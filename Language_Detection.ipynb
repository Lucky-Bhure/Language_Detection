```python
# Split the data into dependent and independent variables
x = np.array(df["Text"])       # independent
y = np.array(df["language"])   # dependent
```

```python
# Importing the required libraries for training the datasheet

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

from sklearn import metrics
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
# Vectorize and fit & transform the data
cv = CountVectorizer()
x = cv.fit_transform(x)
```

```python
# Applying the train_test_split method
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
# Applying NLP algorithm to train the model
model = MultinomialNB()
model.fit(xtrain,ytrain)
```

```python
# Make the pickle to file

import pickle

file = "model.pkl"

fileobj = open(file,'wb')

pickle.dump(model, fileobj)

fileobj.close()
```

```python
# Read the pickle file binary mode
file = "model.pkl"

fileobj = open(file,'rb')

model = pickle.load(fileobj)
```

```python
# Predict the output for previous Text
ypred = model.predict(xtest)
```

```python
# Show the accuracy of model
accuracy_score = metrics.accuracy_score(ytest,ypred)
accuracy_score
```

```python
# Take a user input as a sentence and get a output as language
user_input = input("Enter the Text : ")
data = cv.transform([user_input]).toarray()
output = model.predict(data)
print("Language : ",output)
```
